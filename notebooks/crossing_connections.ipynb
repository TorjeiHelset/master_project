```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
d_list = [0.1, 0.3, 0.6, 0.8]
```

```python
def small_d_prob(x,d):
    if x < d:
        return x / (1-d)
    elif x < 1-d:
        return d / (1-d)
    else:
        return (1-x) / (1-d)
    
small_d_prob = np.vectorize(small_d_prob)

def large_d_prob(x,d):
    if x < 1-d:
        return x / (1-d)
    elif x < d:
        return 1
    else:
        return (1-x) / (1-d)

large_d_prob = np.vectorize(large_d_prob)

def p(x, d):
    if d < 0.5:
        return small_d_prob(x, d)
    else:
        return large_d_prob(x, d)
```

```python
p_vals = [p(x, d) for d in d_list]
```

```python
for value, d in zip(p_vals, d_list):
    plt.plot(x, value, label=f'd={d}')

plt.legend()
```
