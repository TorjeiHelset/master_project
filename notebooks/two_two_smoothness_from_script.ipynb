### Bus Delay

#### 5 Increment

```python
T = 150
N = 3
speed_limits = [40.0, 40.0, 40.0, 40.0]
controls = [[], [], [], []]
cycle = [60.0, 60.0]

increment =  np.linspace(-30, 30, 13)
```

```python
with contextlib.redirect_stdout(None):
    objectives_1, gradients_1 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 0, 0)
    objectives_2, gradients_2 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 1, 0)
    objectives_3, gradients_3 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 2, 0)
    objectives_4, gradients_4 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 3, 0)
    objectives_5, gradients_5 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 4, 0)
    objectives_6, gradients_6 = smooth.two_two_junction_smooth(T, N, speed_limits, controls, cycle, increment, 5, 0)
```

```python
delta_inc = increment[1] - increment[0]
fig, axes = plt.subplots(2, 2, figsize=(10,10))

# Speed 1:
for inc, speed_obj, grad in zip(increment, objectives_1, gradients_1):
    axes[0,0].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[0,0].plot(40+increment, [o.detach() for o in objectives_1])
axes[0,0].set_title("Speed 1")

# Speed 2:
for inc, speed_obj, grad in zip(increment, objectives_2, gradients_2):
    axes[0,1].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[0,1].plot(40+increment, [o.detach() for o in objectives_2])
axes[0,1].set_title("Speed 2")

# Speed 3:
for inc, speed_obj, grad in zip(increment, objectives_3, gradients_3):
    axes[1,0].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[1,0].plot(40+increment, [o.detach() for o in objectives_3])
axes[1,0].set_title("Speed 3")

# Speed 4:
for inc, speed_obj, grad in zip(increment, objectives_4, gradients_4):
    axes[1,1].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[1,1].plot(40+increment, [o.detach() for o in objectives_4])
axes[1,1].set_title("Speed 4")

plt.tight_layout()
tikzplotlib.save("../figurer/two_two_bus_delay_5speed.tex")
```

```python
delta_inc = increment[1] - increment[0]
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Time 1:
for inc, cycle_obj, grad in zip(increment, objectives_5, gradients_5):
    axes[0].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[0].plot(60+increment, [o.detach() for o in objectives_5])
axes[0].set_title("Time 1")

# Time 2:
for inc, cycle_obj, grad in zip(increment, objectives_6, gradients_6):
    axes[1].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[1].plot(60+increment, [o.detach() for o in objectives_6])
axes[1].set_title("Time 2")


plt.tight_layout()
tikzplotlib.save("../figurer/two_two_bus_delay_5time.tex")
```
