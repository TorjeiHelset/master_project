```python
import matplotlib.pyplot as plt
import torch
import numpy as np
```

```python
def flux(rho):
    return rho * (1. - rho)

def d_flux(rho):
    return torch.abs(1. - 2*rho)
```

```python
rho = torch.linspace(0,0.5,1000)
f = flux(rho)
df = d_flux(rho)

plt.plot(np.array(rho), f)
plt.plot(np.array(rho), df)
```
