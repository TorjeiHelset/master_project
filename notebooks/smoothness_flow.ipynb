#### Testing smoothness of the integral of the densities over all roads over all times

Testing for a small network with only two roads and one traffic light connecting the two roads.

```python
# First testing the speed limits
# Parameters for the network in this case:
variable_speeds = [float(i) for i in range(20, 80)]
first_fixed_speed = 30.0
second_fixed_speed = 60.0
cycle_times = [50.0,60.0]
T = 100
```

```python
# Testing the smoothness wrt the first roads speed limit
first_objectives_v1 = sm.test_smoothness_v1(variable_speeds, first_fixed_speed, cycle_times, T)
second_objectives_v1 = sm.test_smoothness_v1(variable_speeds, second_fixed_speed, cycle_times, T)
```

```python
plt.plot(variable_speeds, [o for o in first_objectives_v1], label = 'First fixed speed')
plt.plot(variable_speeds, [o for o in second_objectives_v1], label = 'Second fixed speed')
plt.xlabel('Variable speed')
plt.ylabel('Objective value')
plt.legend()
plt.show()
```

```python
# Testing the smoothness wrt the second roads speed limit
first_objectives_v2 = sm.test_smoothness_v2(first_fixed_speed, variable_speeds, cycle_times, T)
second_objectives_v2 = sm.test_smoothness_v2(second_fixed_speed, variable_speeds, cycle_times, T)
```

```python
plt.plot(variable_speeds, [o for o in first_objectives_v2], label = 'First fixed speed')
plt.plot(variable_speeds, [o for o in second_objectives_v2], label = 'Second fixed speed')
plt.xlabel('Variable speed')
plt.ylabel('Objective value')
plt.legend()
plt.show()
```
