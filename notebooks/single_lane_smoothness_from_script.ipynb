```python
import sys
sys.path.append("../src")
import smoothness as smooth
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib
import contextlib
```

### Bus Delay

#### 5 Increment

```python
T = 100
N = 3
speed_limits = [40.0]
controls = []

increment =  np.linspace(-20, 20, 9)
```

```python
with contextlib.redirect_stdout(None):
    objectives_1, gradients_1 = smooth.single_lane_smooth(T, N, speed_limits, controls, increment, 0)
```

```python
delta_inc = increment[1] - increment[0]

# Speed 1:
for inc, speed_obj, grad in zip(increment, objectives_1, gradients_1):
    plt.plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
plt.plot(40+increment, [o.detach() for o in objectives_1])

tikzplotlib.save("../figurer/single_lane_bus_delay_5speed.tex")
```
