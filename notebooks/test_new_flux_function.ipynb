```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Critical density rho_c
rho_c = 0.6
v_f = 80
rho_max = 1

def flux(rho):
    if rho < rho_c:
        return rho * v_f
    else:
        return rho_c*v_f * (rho_max  - rho) / (rho_max - rho_c)
v_flux = np.vectorize(flux)
```

```python
rho = np.linspace(0.01,0.99, 20)
fluxes = v_flux(rho)
plt.plot(rho, fluxes)
plt.show()
```
