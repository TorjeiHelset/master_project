```python
import sys
sys.path.append("../src")
import smoothness as smooth
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib
import contextlib
```

### Bus Delay

#### 5 Increment

```python
T = 150
N = 3
speed_limits = [40.0, 40.0]
controls = [[], []]
cycle = [60.0, 60.0]

increment =  np.linspace(-20, 20, 9)
```

```python
with contextlib.redirect_stdout(None):
    objectives_1, gradients_1 = smooth.single_junction_smooth(T, N, speed_limits, controls, cycle, increment, 0, 0)
    objectives_2, gradients_2 = smooth.single_junction_smooth(T, N, speed_limits, controls, cycle, increment, 1, 0)
    objectives_3, gradients_3 = smooth.single_junction_smooth(T, N, speed_limits, controls, cycle, increment, 2, 0)
    objectives_4, gradients_4 = smooth.single_junction_smooth(T, N, speed_limits, controls, cycle, increment, 3, 0)
```

```python
delta_inc = increment[1] - increment[0]
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Speed 1:
for inc, speed_obj, grad in zip(increment, objectives_1, gradients_1):
    axes[0].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[0].plot(40+increment, [o.detach() for o in objectives_1])
axes[0].set_title("Speed 1")

# Speed 2:
for inc, speed_obj, grad in zip(increment, objectives_2, gradients_2):
    axes[1].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[1].plot(40+increment, [o.detach() for o in objectives_2])
axes[1].set_title("Speed 2")


plt.tight_layout()
tikzplotlib.save("../figurer/single_junction_bus_delay_5speed.tex")
```

```python
delta_inc = increment[1] - increment[0]
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Time 1:
for inc, cycle_obj, grad in zip(increment, objectives_3, gradients_3):
    axes[0].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[0].plot(60+increment, [o.detach() for o in objectives_3])
axes[0].set_title("Time 1")

# Time 2:
for inc, cycle_obj, grad in zip(increment, objectives_4, gradients_4):
    axes[1].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[1].plot(60+increment, [o.detach() for o in objectives_4])
axes[1].set_title("Time 2")


plt.tight_layout()
tikzplotlib.save("../figurer/single_junction_bus_delay_5time.tex")
```

```python
delta_inc = increment_3[1] - increment_3[0]
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Speed 1:
for inc, speed_obj, grad in zip(increment_3, objectives_1_two, gradients_1_two):
    axes[0].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[0].plot(40+increment_3, [o.detach() for o in objectives_1_two])
axes[0].set_title("Speed 1")

# Speed 2:
for inc, speed_obj, grad in zip(increment_3, objectives_2_two, gradients_2_two):
    axes[1].plot([40 + inc - delta_inc/2, 40 + inc + delta_inc/2], [speed_obj.detach() - grad * delta_inc / 2, speed_obj.detach() + grad * delta_inc / 2], '--r')
axes[1].plot(40+increment_3, [o.detach() for o in objectives_2_two])
axes[1].set_title("Speed 2")

plt.tight_layout()
tikzplotlib.save("../figurer/single_junction_travel_time_01speed.tex")
```

```python
delta_inc = increment_3[1] - increment_3[0]
fig, axes = plt.subplots(1, 2, figsize=(10,5))

# Time 1:
for inc, cycle_obj, grad in zip(increment_3, objectives_3_two, gradients_3_two):
    axes[0].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[0].plot(60+increment_3, [o.detach() for o in objectives_3_two])
axes[0].set_title("Time 1")

# Time 2:
for inc, cycle_obj, grad in zip(increment_3, objectives_4_two, gradients_4_two):
    axes[1].plot([60 + inc - delta_inc/2, 60 + inc + delta_inc/2], [cycle_obj.detach() - grad * delta_inc / 2, cycle_obj.detach() + grad * delta_inc / 2], '--r')
axes[1].plot(60+increment_3, [o.detach() for o in objectives_4_two])
axes[1].set_title("Time 2")


plt.tight_layout()
tikzplotlib.save("../figurer/single_junction_travel_time_01time.tex")
```
