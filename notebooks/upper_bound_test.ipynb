```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d1 = 0.1
d2 = 0.4
```

```python
def upper_bound(d1, d2):
    return 1 / ((1-d1) * (1-d2)) * (-1 + 2/3 *d1**3 + 10 * d1**2*d2 - 25/3 * d1**2 - 15/2 * d1*d2 + 8*d1 - 3*d1*d2**2 + 15/2 * d2**2 - d2)
upper_bound = np.vectorize(upper_bound)
```

```python
x = np.linspace(0, 0.5, 10)
y = np.linspace(0, 0.5, 10)
X, Y = np.meshgrid(x, y)
Z = upper_bound(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111)
im = ax.imshow(Z, cmap='viridis', extent=[0, 0.5, 0, 0.5])
plt.colorbar(im)
plt.show()

# fig = plt.figure()
# ax = fig.add_subplot(111)
# ax.contour(X, Y, Z)
# plt.show()

# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# ax.plot_surface(X, Y, Z)
# plt.show()
```
