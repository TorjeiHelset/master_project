```python
import json

f = open("../results/kvadraturen_500_temp_opt.json")
data = json.load(f)
f.close()
densities = data[0]

f = open("../results/kvadraturen_500_orig.json")
data = json.load(f)
f.close()
lengths = data[2]
```

```python
# Need to convert lengths to have the same time steps as densities does
densities['0'].keys()
```

```python
lengths['0'].keys()
```

```python
new_times = list(densities['0'].keys())
new_times = [float(t) for t in new_times]

old_times = list(lengths['0'].keys())
old_times = [float(t) for t in old_times]
```

```python
new_lengths = {i : {} for i in lengths.keys()}
prev_idx = 0
for t in new_times:
    add_incr = 0
    for old_t_prev, old_t_next in zip(old_times[prev_idx:-1], old_times[prev_idx+1:]):
        if old_t_prev <= t and old_t_next >= t:
            # T somewhere between oldt_t_prev and old_t_next
            prev_dist = t - old_t_prev
            next_dist = old_t_next - t
            interval_length = old_t_next - old_t_prev
            prev_weight = (interval_length - prev_dist) / interval_length
            next_weight = (interval_length - next_dist) / interval_length
            for bus_id in lengths.keys(): 
                new_lengths[bus_id][str(t)] = lengths[bus_id][str(old_t_prev)]*prev_weight + lengths[bus_id][str(old_t_next)]*next_weight
            break
        else:
            add_incr += 1
```

```python
test_times = list(new_lengths['0'].keys())
test_times = [float(t) for t in test_times]
```

```python
with open("kvadraturen_500_orig_lengths.json", 'w') as fd:
    fd.write(json.dumps([new_lengths]))
```
