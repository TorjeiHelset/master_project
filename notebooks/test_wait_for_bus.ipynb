```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid_up(x):
    return 1 / (1 + torch.exp(-x-5))
def sigmoid_down(x):
    return -1 / (1 + torch.exp(-x+5))
```

```python
x = torch.arange(-20, 20, 0.1)
y = np.array((sigmoid_up(0.5*x) + sigmoid_down(2*x))**1)
y_2 = np.array((sigmoid_up(0.5*x) + sigmoid_down(2*x))**2)
```

```python
plt.plot(np.array(x), y)
plt.plot(np.array(x), y_2)

# plt.plot([-10, 10], [0.99, 0.99])
plt.plot([0, 0], [0,1])
```
